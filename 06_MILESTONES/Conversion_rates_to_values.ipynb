```python
import pandas as pd
import numpy as np

from_dir = 'proj_rates/'
to_dir = "proj_values_EU28/"

ds1 = pd.read_csv('european_dataset_1.csv', index_col= ['years'])
ds2 = pd.read_csv('european_dataset_2.csv', index_col= ['years'])

vars17 = [ds1['pop'][2017], ds2['gdp_cap'][2017], ds2['ei_total'][2017], ds1['coal_cons'][2017], ds1['oil_cons'][2017],
          ds1['gas_cons'][2017], ds1['total_cons'][2017], ds1['fossil_emi'][2017]]


pop = ['pop_SSP1.csv', 'pop_SSP2.csv', 'pop_SSP3.csv', 'pop_SSP4.csv', 'pop_SSP5.csv']
gdpc = ['gdp_cap_SSP1.csv', 'gdp_cap_SSP2.csv', 'gdp_cap_SSP3.csv', 'gdp_cap_SSP4.csv', 'gdp_cap_SSP5.csv']
ei = ['EnergyIntensity_SSP1.csv', 'EnergyIntensity_SSP2.csv', 'EnergyIntensity_SSP3.csv', 'EnergyIntensity_SSP4.csv',
      'EnergyIntensity_SSP5.csv']
tot_cons = ['total_cons_SSP1.csv', 'total_cons_SSP2.csv', 'total_cons_SSP3.csv', 'total_cons_SSP4.csv', 'total_cons_SSP5.csv']
coal_cons = ['coal_SSP1.csv', 'coal_SSP2.csv', 'coal_SSP3.csv', 'coal_SSP4.csv', 'coal_SSP5.csv']
oil_cons = ['oil_SSP1.csv', 'oil_SSP2.csv', 'oil_SSP3.csv', 'oil_SSP4.csv', 'oil_SSP5.csv']
gas_cons = ['gas_SSP1.csv', 'gas_SSP2.csv', 'gas_SSP3.csv', 'gas_SSP4.csv', 'gas_SSP5.csv']
co2 = ['raw_co2R_SSP1.csv', 'raw_co2R_SSP2.csv', 'raw_co2R_SSP3.csv', 'raw_co2R_SSP4.csv', 'raw_co2R_SSP5.csv']

rates_var = [pop, gdpc, ei, coal_cons, oil_cons, gas_cons, tot_cons, co2]
for var in np.arange(len(rates_var)):
    for file in rates_var[var]:
        print(file)
        df = pd.read_csv(from_dir+file, index_col=['Unnamed: 0'])
        m = df.to_numpy()
        for i in np.arange(m.shape[0]):
            val = vars17[var]
            for j in np.arange(m.shape[1]):
                val = val*(m[i][j]/100 + 1)
                m[i][j] = val

        df = pd.DataFrame(m)

        print(df)
        df.to_csv(to_dir+file)
```
